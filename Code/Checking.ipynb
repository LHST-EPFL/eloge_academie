```python
ls = list()
for row in sheet.rows:
    try:
        ls.append(row[2].value)
    except:
        pass
```

```python
ls.pop(0)
```

```python
df_eul = eul[eul['Annee']>1800]
list_to_check = list(df_eul.Savant)
list_to_check = [str(i).lower() for i in list_to_check]
ls = [str(i).lower() for i in ls]
```

```python
def check_if_in(string, li = list_to_check):
    for sav in li :
        if string in str(sav) :
            return False
        elif str(sav) in string :
            return False
    return True
```

```python
check_if_in('condorcet')
```

```python
not_in_eul = [i for i in ls if check_if_in(str(i))]
```

```python
not_find = [i for i in list_to_check if check_if_in(str(i), ls)]
```
