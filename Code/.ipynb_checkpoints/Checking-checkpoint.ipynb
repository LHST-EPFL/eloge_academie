```python
import pandas as pd
```

```python
#https://www.marsja.se/your-guide-to-reading-excel-xlsx-files-in-python/

import openpyxl
from pathlib import Path


#importing the file with the list of eulogies
xlsx_file = Path('./../Data', 'Éloges 19e s.xlsx')
wb_obj = openpyxl.load_workbook(xlsx_file) 

# Read the active sheet:
sheet = wb_obj.active
```

```python
eul = pd.read_csv('./../Data/Dataframes/eulogies.csv')
```

```python
df_1 = pd.read_csv('./../Data/Dataframes/dataframe_Condorcet_1.csv')
df_2 = pd.read_csv('./../Data/Dataframes/dataframe_Condorcet_2.csv')

df = pd.concat([df_1, df_2], ignore_index = True)
```

```python
#Printing the values in the xls file

for row in sheet.iter_rows():
    for cell in row:
        print(cell.value, end=" ")
    print()
```

```python
#Getting the value of the column Savant for our list to check
ls = list()
for row in sheet.rows:
    try:
        ls.append(row[2].value)
    except:
        pass
```

```python
#pop first value
ls.pop(0)
```

```python
#Keeping only the eulogies written in the nineteenth century
df_eul = eul[eul['Annee']>1800]
list_to_check = list(df_eul.Savant)
list_to_check = [str(i).lower() for i in list_to_check]
ls = [str(i).lower() for i in ls]
```

```python
def check_if_in(string, li = list_to_check):
    '''
    Method to check if the scientist is present in the list to check.
    Returns False if it is in it, True if the scientist isn't in it
    '''
    for sav in li :
        if string in str(sav) :
            return False
        elif str(sav) in string :
            return False
    return True
```

```python
#Test
check_if_in('condorcet')
```

```python
#List of the eulogies that are not present in our dataset
not_in_eul = [i for i in ls if check_if_in(str(i))]
```

```python
#List of the eulogies in our dataset that are not listed in the xls file
not_find = [i for i in list_to_check if check_if_in(str(i), ls)]
```

```python
#Checking by hand the rest of the list as they are different ways of writing the same name

not_in_eul.pop(not_in_eul.index('pingré, alexandre-guy'))
not_find.pop(not_find.index('beudant et legrand des cloizeaux'))
not_in_eul
```
